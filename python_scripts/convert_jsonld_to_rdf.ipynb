### Verify problem metadata files

```python
from rdflib import Graph
from pathlib import Path
from pyshacl import validate

input_folder = Path("/Users/anakostovska/Dropbox/JSI_active_studies/COMON/data/raw_data/metadata/problem_metadata/")
output_folder = Path("/Users/anakostovska/Dropbox/JSI_active_studies/COMON/data/annotations/RDFannotations/problem_metadata/")
shacl_file = Path("/Users/anakostovska/Dropbox/JSI_active_studies/COMON/shapes/COMON_problem_shapes.ttl")

for file in input_folder.glob("*-LD.json"):
    print(f"\n{'='*50}")
    print(f"Processing: {file.name}")

    # --- Validate with SHACL ---
    conforms, _, results_text = validate(
        data_graph=str(file),
        shacl_graph=str(shacl_file),
        data_graph_format="json-ld",
        shacl_graph_format="turtle",
        inference="rdfs"
    )

    print(f"SHACL Conforms: {conforms}")

    if not conforms:
        print("Validation errors:")
        print(results_text)
        print("Skipping serialization due to validation errors.")
        continue  # skip to next file

    # --- Serialize to RDF/XML only if valid ---
    g = Graph()
    g.parse(file, format="json-ld")

    output_file = output_folder / (file.stem.replace("-LD", "") + ".owl")
    g.serialize(destination=output_file, format="xml")

    print(f"Done. {len(g)} triples written to {output_file}")
```
